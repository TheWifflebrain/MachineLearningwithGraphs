```python
data = []

df1 = pd.read_csv('../twitch/DE/musae_DE_edges.csv')
df2 = pd.read_csv('../twitch/ENGB/musae_ENGB_edges.csv')
df3 = pd.read_csv('../twitch/ES/musae_ES_edges.csv')
df4 = pd.read_csv('../twitch/FR/musae_FR_edges.csv')
df5 = pd.read_csv('../twitch/PTBR/musae_PTBR_edges.csv')
df6 = pd.read_csv('../twitch/RU/musae_RU_edges.csv')

f = open('../twitch/DE/musae_DE.json')
dat = json.load(f)
data.append(dat)

f = open('../twitch/ENGB/musae_ENGB_features.json')
dat = json.load(f)
data.append(dat)

f = open('../twitch/ES/musae_ES_features.json')
dat = json.load(f)
data.append(dat)

f = open('../twitch/FR/musae_FR_features.json')
dat = json.load(f)
data.append(dat)

f = open('../twitch/PTBR/musae_PTBR_features.json')
dat = json.load(f)
data.append(dat)

f = open('../twitch/RU/musae_RU_features.json')
dat = json.load(f)
data.append(dat)

feat = []
for d in data:
    X = []
    longest_len = 0
    shortest_len = 90000
    for key in d:
        value = d[key]
        if len(value) > longest_len: longest_len = len(value)
        if len(value) < shortest_len: shortest_len = len(value)
        value.append(int(key))
        X.append(value)

    print(longest_len)
    print(shortest_len)

    X = sorted(X, key=lambda x:x[-1])
    print(X[1])

    X = [ l[0:len(l)-1] for l in X ]
    X2 = X

    X = [ l[:shortest_len] for l in X ]

    for i, b in enumerate(X2):
        X2[i] = (b + longest_len * [0])[:longest_len]

    # X = truncated for shortest node features
    # X2 = padded to longest node features
    # print(X[1])
    # print(X2[1])
    feat.append([X, X2])

```

Truncate Node Features

```python
import os
os.environ["OMP_NUM_THREADS"] = '8'

titles = ['DE', 'ENG', 'ES', 'FR', 'PTBR', 'RU', 'Large Twitch Network']
com_num = [11,21,12,8,8,13,20]

import warnings
warnings.filterwarnings('ignore')

for count, g in enumerate(GS):
    if(count == 6): break
    start_time = time.time()

    labels_trunc = []
    mod_trunc = []
    par_trunc = []
    # labels_trunc.append([0])
    # mod_trunc.append(0)
    # par_trunc.append(0)

    for k in range(1, 101):
        kmeans_trunc = KMeans(n_clusters=k, random_state=0).fit(feat[count][0])
        labels_trunc.append(kmeans_trunc.labels_)

        k_comms_trunc = [[] for _ in range(k)]
        for i, kl in enumerate(kmeans_trunc.labels_):
            community_num = kl
            k_comms_trunc[kl].append(i)

        for i, kl in enumerate(k_comms_trunc):
            k_comms_trunc[i] = set(kl)

        k_mod_trunc = nx_comm.modularity(GS[count], k_comms_trunc)
        mod_trunc.append(k_mod_trunc)
        k_par_trunc = nx_comm.partition_quality(GS[count], k_comms_trunc)
        par_trunc.append(k_par_trunc)

    print("--- %s seconds ---" % (time.time() - start_time))

    max_value_trunc = max(mod_trunc)
    max_index_trunc = mod_trunc.index(max_value_trunc)+1
    print("Max Value: ", max_value_trunc)
    print("Max Index: ", max_index_trunc)

    min_value_trunc = min(mod_trunc)
    min_index_trunc = mod_trunc.index(min_value_trunc)+1
    print("Min Value: ", min_value_trunc)
    print("Min Index: ", min_index_trunc)

    k_num_trunc = list(range(1, 101))
  
    plt.plot(k_num_trunc, mod_trunc, color='blue', marker='o')
    plt.title(f'{titles[count]}: Kmeans Truncated', fontsize=14)
    plt.xlabel('Number of Communities', fontsize=14)
    plt.ylabel('Modularity Score', fontsize=14)
    plt.grid(True)
    plt.show()
```

```python
# from numpy.linalg import eig
# from scipy.sparse import csgraph
# import numpy as np
# from sklearn.cluster import SpectralClustering
# from sklearn import metrics
coms = [11,21,12,8,8,13,20]
for count, g in enumerate(GS):
    start_time = time.time()
    labels_s = []
    mod_s = []
    par_s = []

    # labels_s.append([0])
    # mod_s.append(0)
    # par_s.append(0)

    A = nx.adjacency_matrix(g)
    A = A.toarray()
    for k in range(1, 31):
        if k % 10 == 0: print(k)
        sc = SpectralClustering(k, affinity='precomputed', n_init=100)
        sc.fit(A)

        labels_s.append(sc.labels_)

        cm_comms = [[] for _ in range(k)]
        for i,kl in enumerate(sc.labels_):
            community_num = kl
            cm_comms[kl].append(i)

        for i,kl in enumerate(cm_comms):
            cm_comms[i] = set(kl)

        s_mod = nx_comm.modularity(GS[count], cm_comms)
        mod_s.append(s_mod)
        s_par = nx_comm.partition_quality(GS[count], cm_comms)
        par_s.append(s_par)

    print("--- %s seconds ---" % (time.time() - start_time))

    max_value_s = max(mod_s)
    max_index_s = mod_s.index(max_value_s)+1
    print("Max Value: ", max_value_s)
    print("Max Index: ", max_index_s)

    min_value_s = min(mod_s)
    min_index_s = mod_s.index(min_value_s)+1
    print("Min Value: ", min_value_s)
    print("Min Index: ", min_index_s)

    sc_l = list(range(1,31))

    plt.plot(sc_l, mod_s, color='blue', marker='o')
    plt.title(f'{titles[count]}: Spectral Clustering', fontsize=14)
    plt.xlabel('Number of Communities', fontsize=14)
    plt.ylabel('Modularity Score', fontsize=14)
    plt.grid(True)
    plt.show()
```

```python
# from homework assignment from professor
from node2vec import Node2Vec
import networkx as nx

class Node2Vec(Node2Vec):
  """
  Parameters
  ----------
  p : float
      p parameter of node2vec
  q : float
      q parameter of node2vec
  d : int
      dimensionality of the embedding vectors
  """
  def __init__(self, graph, p=1, q=1, d=32):
    super().__init__(
                     graph = graph,
                     walk_length=10,
                     p=p,
                     q=q,
                     dimensions =d
                  )

KS = []
for count, g in enumerate(GS):
    start_time = time.time()
    n2v_model = Node2Vec(GS[count], 1, 1, 16)

    # Embed nodes
    model = n2v_model.fit(window=10, min_count=1, batch_words=4)

    # Node to vec representation
    K16 = model.wv.vectors
    K16.shape

    print(K16[0])
    print(type(K16))
    KS.append(K16)
    print("--- %s seconds ---" % (time.time() - start_time))
```
