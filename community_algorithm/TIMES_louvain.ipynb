Louvain Comm


```python
for g in GS:
    start_time = time.time()
    louvain = nx_comm.louvain_communities(g)
    num_communities = len(louvain)
    size_community = [len(c) for c in louvain]
    print("--- %s seconds ---" % (time.time() - start_time))
    print(num_communities)
    print(size_community)
    print(np.sum(size_community))

    l_mod = nx_comm.modularity(g, louvain)
    print('modularity:',l_mod)
    print('\n\n')
```

greedy_modularity_communities

```python
# take like 90 secs
for count, g in enumerate(GS):
    if(count == 6): break
    start_time = time.time()
    gm = greedy_modularity_communities(g)

    print("--- %s seconds ---" % (time.time() - start_time))
    
    num_communities = len(gm)
    size_community = [len(c) for c in gm]
    print(num_communities)
    print(size_community)

    gm_mod = nx_comm.modularity(g, gm)
    print('modularity:',gm_mod)
    print('\n\n')
```
